```python
import numpy as np
import scipy as sp
import cv2
import matplotlib.pyplot as plt
import sys
%matplotlib inline
```

```python
def to_gray(image):
    img = image.copy()
    gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return gray_image
```

```python
output_image = cv2.imread('img1.jpg')
image_2 = cv2.imread('img2.jpg')
min_xy = np.array([-995.40045166, -409.15112305])
```

```python
min_xy = min_xy.astype(np.int)
min_xy
```

```python
# plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
output_image = cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB)
plt.imshow(output_image)
```

```python
output_image2 = np.zeros(output_image.shape, dtype=np.uint8)
output_image2[-min_xy[1]:-min_xy[1] + image_2.shape[0],
              -min_xy[0]:-min_xy[0] + image_2.shape[1]] = image_2
output_image2 = cv2.cvtColor(output_image2, cv2.COLOR_BGR2RGB)
# plt.imshow(cv2.cvtColor(output_image2, cv2.COLOR_BGR2RGB))
plt.imshow(output_image2)
```

```python
img2_map = np.zeros((output_image.shape[0], output_image.shape[1]))
img2_map[-min_xy[1]:-min_xy[1] + image_2.shape[0],
         -min_xy[0]:-min_xy[0] + image_2.shape[1]] = 1
plt.imshow(img2_map, cmap='gray')
```

```python
img1_map = np.zeros((output_image.shape[0], output_image.shape[1]))
img1_map[to_gray(output_image) != 0] = 1
plt.imshow(img1_map, cmap='gray')
```

```python
overlap_map = np.zeros((output_image.shape[0], output_image.shape[1]))
overlap_map[np.logical_and(img1_map == 1, img2_map == 1)] = 1
plt.imshow(overlap_map, cmap='gray')
```

```python
plt.imshow(1-(img1_map-overlap_map), cmap='gray')
```

```python
from scipy import ndimage
```

```python
plt.imshow(1-(img2_map-overlap_map), cmap='gray')
```

```python
d1 = ndimage.distance_transform_edt(1-(img1_map-overlap_map))
d2 = ndimage.distance_transform_edt(1-(img2_map-overlap_map))
gradient = d1.copy()
gradient = gradient.astype(np.float32) / (d1+d2)
gradient[overlap_map == 0] = 0
plt.imshow(gradient, cmap='gray')
```

```python
mask = img1_map.copy()
mask[overlap_map == 1] = (1-gradient)[overlap_map == 1]
plt.imshow(mask, cmap='gray')
```

```python
def alpha_blend(img1, img2, mask):
    blend = np.zeros(img1.shape, dtype=np.uint8)
    if len(img1.shape) == 2:
        blend = img1 * mask + img2 * (1-mask)
    else:
        for ch in range(3):
            blend[:,:,ch] = img1[:,:,ch] * mask + img2[:,:,ch] * (1-mask)
    return blend
```

```python
blend = alpha_blend(to_gray(output_image), to_gray(output_image2), mask)
plt.imshow(blend, cmap='gray')
```
